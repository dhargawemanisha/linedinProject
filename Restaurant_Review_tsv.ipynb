```python
import pandas as pd
```

```python
df= pd.read_csv("/content/Restaurant_Reviews.tsv",delimiter='\t')
```

```python
df.head()
```

```python
df.shape
```

```python
import re
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
nltk.download("stopwords")

corpus = []
for i in range(0,1000):
  ## Remove Punctuation
  review =re.sub("[^a-zA-z | ^\w+'t]",' ',df["Review"][i])
   ## Convert to lower
  review = review.lower() 

  ## Covert to list of words
  review = review.split(" ") 

  ##Stemming with removing stopwords
  stemmer = PorterStemmer()
            
  review = [stemmer.stem(word) for word in review if not word in stopwords.words('english')]

  ## Convert back to string from list
  review = " ".join(review)
  corpus.append(review)
  print(df["Review"][i], "==>", review)
   
```

```python
print (corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

x = cv.fit_transform(corpus).toarray()
```

```python
df.head()
```

```python
y= df.iloc[:,1].values
#y=df["Liked"].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.2)
  
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# LogisticRegression
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x_train,y_train)
```

```python
y_pred_test_log = log_model.predict(x_test)
y_pred_test_log
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_test_log))
```

```python
from sklearn.naive_bayes import GaussianNB
model_NB =GaussianNB()
model_NB.fit(x_train,y_train)
y_pred_test_NB =model_NB.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_test_NB))
```

```python
from sklearn.svm import SVC
model_SVM = SVC(kernel="rbf")
model_SVM.fit(x_train,y_train)
y_pred_test_SVM = model_SVM.predict(x_test)
```

```python
print(classification_report(y_test,y_pred_test_SVM))
```
